```python
from Bio import Phylo
from Bio.Phylo.Consensus import majority_consensus
```

```python
with open('main_complete_orthologs_renamed_to_nwk.tsv', 'r') as f:
    filenames = [line.split('\t')[0].strip() for line in f]
```

```python
print(filenames)
```

```python
import os

trees = []

for filename in filenames:
    if os.path.isfile(filename):
        trees.append(Phylo.read(filename, 'newick'))
    else:
        print(f"filenotfound  : {filename}")
```

```python
consensus_tree = majority_consensus(trees, 0.5)
```

```python
Phylo.write(consensus_tree, 'consensus_main_new.nwk', 'newick')
```
