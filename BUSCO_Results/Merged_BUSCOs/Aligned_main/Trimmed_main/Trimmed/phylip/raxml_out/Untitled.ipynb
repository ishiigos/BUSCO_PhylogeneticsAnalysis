```python
from Bio import Phylo
from Bio.Phylo.Consensus import majority_consensus
```

```python
with open('common_BUSCOs_main_species.tsv', 'r') as f:
    filenames = [line.split('\t')[0].strip() for line in f]
```

```python
print(filenames)
```

```python
import os

trees = []
file = "RAxML_bipartitions."
end = "_trimmed.nwk"
for filename in filenames:
    filename = file+filename+end
    if os.path.isfile(filename):
        trees.append(Phylo.read(filename, 'newick'))
    else:
        print(f"filenotfound  : {filename}")
```

```python
consensus_tree = majority_consensus(trees)
```

```python
Phylo.write(consensus_tree, 'consensus_main_bipartition.nwk', 'newick')
```
